##Step 1: Load and Explore the Dataset

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline

# Load the dataset (after unzipping manually in the left panel of Colab)
df = pd.read_csv('https://github.com/FourthRevGitHub/CAM_DS_Public-Data-Sets/raw/main/Course%201/Week%206/CUSTOMERS_CLEAN.zip')

# Preview the dataset
df.head()
```

##Step 2: Basic Dataset Information

```python
# Check the structure of the dataset
df.info()

# Check for missing values
df.isnull().sum()

# Check for duplicates
df.duplicated().sum()
```

##Step 3: Feature Engineering — Frequency, Recency, CLV, Avg Unit Cost, Age

```python
# Convert dates to datetime
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], errors='coerce')
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Customer_BirthDate'] = pd.to_datetime(df['Customer_BirthDate'], errors='coerce')

# Convert revenue and cost to numeric (remove $ and commas if needed)
df['Total Revenue'] = df['Total Revenue'].replace('[\$,]', '', regex=True).astype(float)
df['Unit Cost'] = df['Unit Cost'].replace('[\$,]', '', regex=True).astype(float)

# ➤ FREQUENCY: Number of orders per customer
df['Frequency'] = df.groupby('Customer ID')['Order ID'].transform('count')

# ➤ RECENCY: Days since last delivery
df['Recency'] = (pd.to_datetime('today') - df['Delivery_Date']).dt.days

# ➤ CLV: Equal to total revenue
df.rename(columns={'Total Revenue': 'CLV'}, inplace=True)

# ➤ Average Unit Cost (rename for clarity)
df.rename(columns={'Unit Cost': 'Avg_Unit_Cost'}, inplace=True)

# ➤ Customer AGE
today = pd.Timestamp('now')
df['Age'] = (today - df['Customer_BirthDate']).dt.days // 365

# Check new columns
df[['Customer ID', 'Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Age']].head()
```

##Step 4: Aggregate by Customer ID

```python
# Select only relevant columns for aggregation
selected_columns = ['Customer ID', 'Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Age']
df_selected = df[selected_columns]

# Group by Customer ID and keep the first value of each metric
df_grouped = df_selected.groupby('Customer ID').agg({
    'Frequency': 'first',
    'Recency': 'first',
    'CLV': 'sum',  # total value contributed
    'Avg_Unit_Cost': 'mean',
    'Age': 'first'
}).reset_index()

# Preview the new dataset
df_grouped.head()
```

##Step 5: Exploratory Data Analysis (EDA) – Visual Check

```python
# Set up the style
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')

# Histograms of each feature
df_grouped.hist(bins=30, figsize=(15,10), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Features", fontsize=16)
plt.tight_layout()
plt.show()
```

##Step 6: Scaling and Finding Optimal k

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Select only numerical features to scale
features_to_scale = ['Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Age']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_grouped[features_to_scale])

# Create Elbow plot
inertia = []
silhouette_scores = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot Elbow Method
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')

# Plot Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'go-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.tight_layout()
plt.show()
```

```python
# Apply KMeans with the chosen number of clusters
k_optimal = 5
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
df_grouped['Cluster'] = kmeans.fit_predict(X_scaled)

# Preview the result
df_grouped[['Customer ID', 'Cluster']].head()

# Combine the clustering output with customer ID and the 5 engineered features
df_final = df_grouped[['Customer ID', 'Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Age', 'Cluster']].copy()

# Preview the final dataframe
df_final.head()
```

##Step 7: Boxplots by Cluster

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine scaled features and cluster labels into one DataFrame
df_scaled = pd.DataFrame(X_scaled, columns=features_to_scale)
df_scaled['Cluster'] = df_grouped['Cluster'].values

# Plot boxplots for each feature by cluster
plt.figure(figsize=(18, 12))

for i, feature in enumerate(features_to_scale):
    plt.subplot(2, 3, i+1)
    sns.boxplot(data=df_scaled, x='Cluster', y=feature, palette='Set2')
    plt.title(f'{feature} by Cluster')

plt.tight_layout()
plt.show()
```

##Step 8: Hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Sample a subset to avoid crashing
sample_df = pd.DataFrame(X_scaled, columns=features_to_scale).sample(n=3000, random_state=42)

# Compute linkage
Z = linkage(sample_df, method='ward')

# Plot dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=10, leaf_rotation=90, leaf_font_size=12)
plt.title('Dendrogram of Customer Clusters (truncated)')
plt.xlabel('Cluster Size or Index')
plt.ylabel('Distance')
plt.show()
```

##Step 9: PCA and t-SNE Visualization

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# PCA - 2 components
pca = PCA(n_components=2, random_state=42)
pca_result = pca.fit_transform(X_scaled)
df_grouped['PCA1'] = pca_result[:, 0]
df_grouped['PCA2'] = pca_result[:, 1]

# Plot PCA result
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_grouped, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title('PCA - Customer Segments')
plt.show()

# t-SNE - 2 components
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
tsne_result = tsne.fit_transform(X_scaled)
df_grouped['TSNE1'] = tsne_result[:, 0]
df_grouped['TSNE2'] = tsne_result[:, 1]

# Plot t-SNE result
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_grouped, x='TSNE1', y='TSNE2', hue='Cluster', palette='Set2')
plt.title('t-SNE - Customer Segments')
plt.show()
```
